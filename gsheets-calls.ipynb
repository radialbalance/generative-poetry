## Exploring Dictionaries & Lists <a class="anchor" id="dictionaries"></a>


```python
import random
import requests
from io import StringIO
import csv

formURL = "https://docs.google.com/spreadsheets/d/e/2PACX-1vQI4Cb2qL79wKP_Rv34DCGErzO1e4vJiAaLzI2grrQ-9c9rgZGRGxsyB0rarho4iffNtaVp-5T7mOBA/pub?gid=2016707830&single=true&output=csv"

response = requests.get(formURL)

stringResponse = response.content.decode("utf-8")

print(stringResponse)
```

```python
csvFromString = StringIO(stringResponse)

readCSV = csv.DictReader(csvFromString,delimiter=",")

feeling = []
place = []
unknown = []
past = []
future = []

allCells = []

for eachRow in readCSV:
    for eachCol, eachVal in eachRow.items():
        #print(eachCol, eachVal)
        
        allCells.append(eachVal)
        
        if eachCol == "Feeling":
            feeling.append(eachVal)
            
        elif eachCol == "Place":
            place.append(eachVal)
        
        elif eachCol == "Unknown":
            unknown.append(eachVal)
            
        elif eachCol == "Past":
            past.append(eachVal)
            
        elif eachCol == "Future":
            future.append(eachVal)
        
        else:
            pass

print(feeling)
print(place)
print(unknown)
print(past)
print(future)

print(allCells)

random.shuffle(allCells)
print(allCells)
```

```python
listOfLists=[feeling, place, unknown, past, future]

print(listOfLists)
```

# Exploring Dictionaries & Lists 2.0<a class="anchor" id="dictionaries2"></a>


```python
import random
import requests
from io import StringIO
import csv

formURL = "https://docs.google.com/spreadsheets/d/e/2PACX-1vQI4Cb2qL79wKP_Rv34DCGErzO1e4vJiAaLzI2grrQ-9c9rgZGRGxsyB0rarho4iffNtaVp-5T7mOBA/pub?gid=1988907075&single=true&output=csv"

response = requests.get(formURL)

stringResponse = response.content.decode("utf-8")

print(stringResponse)
```

```python
csvFromString = StringIO(stringResponse)

readCSV = csv.DictReader(csvFromString,delimiter=",")

allCells = []

spanish = []

english = []

for eachRow in readCSV:
    for eachCol, eachVal in eachRow.items():
        if eachCol == "Spanish":
            spanish.append(eachVal)
        elif eachCol == "English":
            english.append(eachVal)
            
        #print(eachCol, eachVal)
        
        allCells.append(eachVal)
        
print(allCells)
print(" - "*10)
print(spanish)
print(" - "*10)
print(english)
```

```python
#randomness

for i in range(5):
    selected = random.sample(allCells, 2)
    joined = ' '.join(selected)
    print(joined)
```

```python
count =2
repitions = 4
poemList = []

for line in range(repitions):
    #print(f"[{line*count}:{((line*count)+count)-1}]")
    
    poemRange = allCells[line*count:((line*count)+count)-1]
    
    #print(poemRange)
    
    poemList.append(poemRange)
    
print(poemList)
```

```python
for pair in poemList:
    for poemWords in pair:
        print(poemWords)
```

```python
for pair in poemList:
    for poemWords in enumerate(pair):
        print(poemWords)
```

```python
#weird thing

for pair in poemList:
    for poemWords in enumerate(pair):
        if poemWords[0] == 0:
            print(poemWords[1])
```

```python
for eachLine in enumerate(spanish):
    print(eachLine)
```

## Generative Poetry<a class="anchor" id="poetry-1"></a>


```python
import random
import requests
from io import StringIO
import csv

formURL = "https://docs.google.com/spreadsheets/d/e/2PACX-1vQI4Cb2qL79wKP_Rv34DCGErzO1e4vJiAaLzI2grrQ-9c9rgZGRGxsyB0rarho4iffNtaVp-5T7mOBA/pub?gid=1588715906&single=true&output=csv"

response = requests.get(formURL)

stringResponse = response.content.decode("utf-8")

print(stringResponse)
```

```python
csvFromString = StringIO(stringResponse)

readCSV = csv.DictReader(csvFromString,delimiter=",")
headerRow = readCSV.fieldnames 
csvData = {}

#Identifying the Header Row for our New Data Struct

for eachHeader in headerRow:
    #print(eachHeader)
    csvData[eachHeader] = []

#Appending the values in each Column to the Corresponding Header
#Also checking for empty cells, so they aren't printed

for eachRow in readCSV:
    #print(eachRow)
    for eachCol, eachVal in eachRow.items():
        if eachVal == '':
            pass
        else:
            csvData.get(eachCol).append(eachVal)

#Printing Format for each Header by Name
print(csvData.get('Name'))
```
